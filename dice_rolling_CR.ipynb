```python
import random

dices = 11000
sets = 10000
def roll_set(dices):
    results = []
    for _ in range(dices):
        roll = random.randint(1,20)
        results.append(roll)
    return results

def set_of_sets(sets, dices):
    results = []
    for _ in range(sets):
        results.append(roll_set(dices))
    return results

results = set_of_sets(sets, dices)
```

```python
results_20 = []
for i in results:
    results_20.append(i.count(20))
    
ev = dices * .05
results_ev = []
for i in results_20:
    results_ev.append((i-ev)/i)
```

```python
import statistics as stt
stt.stdev(results_ev)
```

```python
import matplotlib.pyplot as plt
import numpy as np
n, bins, patches = plt.hist(x=results_ev, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.rcParams['figure.figsize'] = [10, 5]
plt.title('Difference vs Expected Value - 10k sets of 11k rolls')
plt.xlabel('Difference')
plt.ylabel('Frequency')
```
